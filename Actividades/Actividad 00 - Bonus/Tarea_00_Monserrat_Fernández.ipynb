```python
# contar valores no recepcionados
len(dfpresupuesto[dfpresupuesto['PRESUPUESTO'] == 'No Recepcionado'])
```

PREGUNTA 2

```python
#Entrega el presupuesto total por región y por provincia, ordenado en cada caso de mayor a menor
```

```python
# realizamos el join de los dataframes que tienen la info del nombre de la región y de la provincia
df1 = pd.merge(left = dfpresupuesto, right = dfcodigos, left_on = 'CODIGO', right_on = 'Código Comuna 2018')
df1.head()
```

```python
# Ordenamos de mayor a menor las regiones según su presupuesto
df1_new = df1.groupby(df1['Nombre Región']).agg({"PRESUPUESTO": 'sum'})
df1_new.sort_values(by='PRESUPUESTO', ascending=False)
```

```python
# Ordenamos de mayor a menor las provincias según su presupuesto
df2_new = df1.groupby(df1['Nombre Provincia']).agg({"PRESUPUESTO": 'sum'})
df2_new.sort_values(by='PRESUPUESTO', ascending=False)
```

PREGUNTA 3

```python
#Entrega el presupuesto promedio por región y por provincia, ordenado en cada caso de mayor a menor. ¿Cómo cambia el ranking cuando se considera el presupuesto total vs. el promedio?
```

```python
# realizamos el join de los dataframes de presupuesto y codigos, que tienen la info del nombre de la región y de la provincia
df1 = pd.merge(left = dfpresupuesto, right = dfcodigos, left_on = 'CODIGO', right_on = 'Código Comuna 2018')
df1.head()
```

```python
# Ordenamos de mayor a menor las regiones según su presupuesto
df1_new2 = df1.groupby(df1['Nombre Región']).agg({"PRESUPUESTO": 'mean'})
df1_new2.sort_values(by='PRESUPUESTO', ascending=False)
```

```python
# Ordenamos de mayor a menor las provincias según su presupuesto
df2_new2 = df1.groupby(df1['Nombre Provincia']).agg({"PRESUPUESTO": 'mean'})
df2_new2.sort_values(by='PRESUPUESTO', ascending=False)
```

**Cuando se comparan los ranking entre el presupuesto total vs el presupuesto promedio, ya sea por región o por provincia, son muy distintos y no están ordenados de la misma manera, por ejemplo, en presupuesto total por región las primeras 3 son metropolitana, valparaíso y biobio y en presupuesto promedio por región son metropolitana, antofagasta y tarapacá. 
Por otro lado, en el presupuesto total por provincia los primeros 3 son Santiago, Valparaíso y Concepción y en el presupuesto promedio por provincia son Santiago, Iquique y Cordillera.**

PREGUNTA 4

```python
#Entrega el top 20 de las comunas con más metros de plaza y el top 20 de las comunas con más presupuesto. ¿Hay alguna coincidencia entre ambos conjuntos?
```

```python
# realizamos el join de los dataframes de presupuesto y comunas, que tienen la info de las comunas
df2 = pd.merge(left = dfpresupuesto, right = dfcomunas, left_on = 'CODIGO', right_on = 'id')
df2.head()
```

```python
df2.fillna(0).head()
```

```python
# obtenemos top 20 de comunas con más metros plaza
df3 = df2.sort_values(by='metros_plaza', ascending=False).head(n=20)
df3[['MUNICIPIO','metros_plaza']]
```

```python
# obtenemos top 20 de comunas con más presupuesto
df4 = df2.sort_values(by='PRESUPUESTO', ascending=False).head(n=20)
df4[['MUNICIPIO','PRESUPUESTO']]
```

```python
# unimos ambos top respecto al código de la comuna, para ver cuales son las coincidencias entre ambos
df5 = pd.merge(left = df3, right = df4, left_on = 'CODIGO', right_on = 'id')
df5['MUNICIPIO_x'].to_frame()
```

**Podemo observar que hay 10 comunas que se repiten en el top 20 de comunas que tienen más metros plaza y más presupuesto y que se muestran en el df5**
